```python


import argparse
import sys
from pathlib import Path

import soundfile as sf
import sherpa_onnx




def create_punct():


    model_dir = Path("./Punct-CT-Transformer/sherpa-onnx-punct-ct-transformer-zh-en-vocab272727-2024-04-12")
    model = model_dir / "model.onnx"

    config = sherpa_onnx.OfflinePunctuationConfig(
        model=sherpa_onnx.OfflinePunctuationModelConfig(ct_transformer=model.as_posix()),
    )
    punct = sherpa_onnx.OfflinePunctuation(config)

    return punct

```

```python
punct = create_punct()
```

```python

text_list = [
    "这是一个测试你好吗How are you我很好thank you are you ok谢谢你",
    "我们都是木头人不会说话不会动",
    "The African blogosphere is rapidly expanding bringing more voices online in the form of commentaries opinions analyses rants and poetry",
]
for text in text_list:
    text_with_punct = punct.add_punctuation(text)
    print("----------")
    print(f"input: {text}")
    print(f"output: {text_with_punct}")

print("----------")
```
